# COVID-19 dataset

This is a dataset derived from the cdc-website containing only laboratory confirmed cases of covid in subjects aged 18 and above(adults) and resident in Maryland, from 2019 to 2023

```python
#importing dependencies
import pandas as pd
import numpy as np
pd.set_option('max_colwidth', 400)
```

## Extracting the data

```python
# Reading the data into a Pandas DataFrame
covid_df = pd.read_csv('Resources\COVID-19_Case_Surveillance_Public_Use_Data_with_Geography_20240613.csv')
covid_df.head()
```

```python
# Get a brief summary of the DataFrame.
covid_df.info()
```

```python
#removing unneccessary columns
covid_df = covid_df.drop(columns =["case_positive_specimen_interval","case_onset_interval","process","current_status","state_fips_code"])
covid_df.head()
```

```python
covid = covid_df.dropna(how = "any")
covid.head()
```

```python
#checking characteristics of remaining columns
covid.info()
```

```python
#splitting date into month and year
#covid['case_month'] = pd.to_datetime(covid['case_month'])

# extracting the month and year components
covid['month'] = covid['case_month'].dt.month
covid['year'] = covid['case_month'].dt.year

covid.head()
```

```python
#determining numberof uunique ethnicities
covid["race"].value_counts()
```

```python
race_df = pd.DataFrame(covid.groupby("race").count())
race_df
```

```python
#determing which county had the highest number of cases
county_df = pd.DataFrame(covid.groupby("res_county")["case_month"].count())
county_df_sorted = pd.DataFrame(county_df["case_month"].sort_values(ascending = False))
county_final = county_df_sorted.rename(columns ={"case_month":"case_count"})
county_final
```

```python
year_df = pd.DataFrame(covid.groupby("year")["case_month"].count())
```
